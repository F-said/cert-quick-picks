## Extract & Transform

Noticing that the first two datasets DO NOT have date, fairly useless. Let's focus in on the last two "gsearch_jobs" and "linkedin-jobs-usa."

Let's join them together and remove any duplicate jobs (assuming that multiple roles ARE NOT posted as multiple postings, but instead as one).

We'll define a duplicate job as duplicate ("title" + "company_name" + "description") and "via" == "via LinkedIn". 

In addition, let's engineer some columns for degree-required, schedule type, and level of job for the last two datasets "gsearch_jobs" and "linkedin-jobs-usa."

Lastly, since "linkedin-jobs-usa" does not have a tech-tokens, we will engineer 2 new columns: one for `tech-tokens` and one for `degree-tokens`. 

In addition, I will also create `level-employment` and `schedule-type`, which will describe what level position and what kind of schedule a position entails.

```python
# prepare jusa for joining by dropping irrevelant cols
jusa_dropped = jusa.drop(columns=["salary", "link"])
```

```python
# prepare jusa for joining by engineering cols

# engineer a column to determine if job is remote 
jusa_dropped["work_from_home"] = np.where(jusa_dropped["onsite_remote"] == "remote", True, False)

jusa_dropped.head()
```

```python
# list of grammar notations
uchars = ",\'\";:-.?!()&/\\[]"

# what cloud compute managed service is mentioned 
clouds = {"azure", "aws", "ibm", "google cloud"}

# include tech tools for fun
techies = {"python", "sql", "excel", "vba", "r", "julia", "pandas", "numpy", "scala", "clojure"}

# degree required
degrees = {"bachelors", "bachelor", "baccalaureate", "masters", "doctorate", "phd"}

# prototype code

# looping is an anti-pattern
for i in range(len(jusa_dropped)):
    # this is an anti-pattern! break up
    # s = set([token.strip(uchars) for token in jusa_dropped["description"].iloc[i].lower().split()]).intersection(clouds)

    row_val = jusa_dropped["description"].iloc[i].lower().split()

    s = set([token.strip(uchars) for token in row_val]).intersection(clouds)

    if len(s) != 0:
        print(s)
```

```python
# split description text by spaces, and get intersection of tools used

cloud_tokens = (
    jusa_dropped["description"].
        apply(lambda x: 
            set([token.strip(uchars) for token in x.lower().split()]).
            intersection(clouds)
        )
)

cloud_tokens.value_counts()
```

```python
tech_tokens = (
    jusa_dropped["description"].
        apply(lambda x: 
            set([token.strip(uchars) for token in x.lower().split()]).
            intersection(techies)
        )
)

tech_tokens.value_counts()
```

```python
degree_tokens = jusa_dropped["description"].apply(lambda x: set([token.strip(uchars) for token in x.lower().split()]).intersection(degrees))

degree_tokens.value_counts()
```

```python
# combine tokens into one dataframe
combined_tokens = pd.concat([cloud_tokens, tech_tokens], axis=1)

combined_tokens.columns = ["desc_cert", "desc_tool"]

combined_tokens
```

```python
# convert all sets to lists
combined_lists = combined_tokens.applymap(lambda x: list(x))
combined_lists
```

```python
# combine all lists together
jusa_dropped["tech_tokens"] = combined_lists["desc_cert"] + combined_lists["desc_tool"]
jusa_dropped["deg_tokens"] = degree_tokens.apply(lambda x: list(x))

jusa_dropped.head()
```

```python
jusa_dropped["criteria"].iloc[0]
```

```python
# transform column into lists
list_crit = jusa_dropped["criteria"].apply(eval).tolist()

# join together all dicts into new_list
res_list = []
for row in list_crit:
    res_dict = {}
    for obj in row:
        res_dict |= obj
    res_list.append(res_dict)

# whoevers bright idea it was to have a list of dictionaries in a column...

# convert into dataframe
df_crit = pd.DataFrame(res_list)
df_crit.head()
```

```python
df_crit["Seniority level"].value_counts()
```

```python
# save information on schedule type and level employment

jusa_dropped["level"] = df_crit["Seniority level"]
jusa_dropped["etype"] = df_crit["Employment type"]
```

```python
# prepare jusa for joining with `gsearch-jobs` by renaming cols
jusa_dropped.rename(columns={"company": "company_name", "posted_date": "date_time"}, inplace=True)

# and then dropping more cols
jusa_dropped.drop(columns=["description", "onsite_remote", "criteria"], inplace=True)

jusa_dropped.head()
```

```python
# after dropping cols, there is a risk for duplicates again, so drop

hash_cols = ["title", "company_name", "location", "date_time", "work_from_home", "level", "etype"]

jusa_dropped.duplicated(subset=hash_cols).sum()
```

```python
jusa_dropped.drop_duplicates(subset=hash_cols, inplace=True)
```

```python
# drop columns for gsearch-jobs
lda_dropped = lda.drop(columns=['Unnamed: 0', 'index', 'via', 'job_id', 'thumbnail', 'posted_at', 'salary', 'search_term', 'search_location', 'commute_time', 'salary_pay', 'salary_rate', 'salary_avg', 'salary_min', 'salary_max', 'salary_hourly', 'salary_yearly', 'salary_standardized'])

lda_dropped.head()
```

```python
# make basic transformations, rename columns and replace Nans in work_from_home

lda_dropped["work_from_home"].fillna(False, inplace=True)

lda_dropped.rename(columns={"description_tokens": "tech_tokens", "schedule_type": "etype"}, inplace=True)
```

```python
# explore extensions & description more
lda_dropped["description"].iloc[0]
```

```python
lda_dropped["extensions"].iloc[100]
```

```python
# extract degree-pertinent tokens
deglda_tokens = (
    lda_dropped["description"].
        apply(lambda x: 
            list(
                set(
                    [token.strip(uchars) for token in x.lower().split()]).intersection(degrees)
                )
            )
)

deglda_tokens.value_counts()
```

```python
levels = {"entry", "mid", "senior", "associate", "executive"}

# extract seniority level of position 
senior_toks = (
    lda_dropped["description"].
        apply(lambda x: 
            list(
                set(
                    [token.strip(uchars) for token in x.lower().split()]).intersection(levels)
                )
            )
)

senior_toks.value_counts()
```

```python
# add back into df
lda_dropped["level"] = senior_toks
lda_dropped["deg_tokens"] = deglda_tokens
```

```python
# drop transformed columns
lda_dropped.drop(columns=["description", "extensions"], inplace=True)
```

```python
test = lda_dropped.copy()

narr = test["level"].to_numpy()

narr
```

```python
# transform "level" into something that contains the "lowest" level employment

# find all non-applicable
lda_dropped["level"] = np.where([x == [] for x in test["level"].to_numpy()], "Not Applicable", lda_dropped["level"])

# find all entry
lda_dropped["level"] = np.where(["entry" in x for x in narr], "Entry level", lda_dropped["level"])

# find all associate
lda_dropped["level"] = np.where(["associate" in x for x in narr], "Associate level", lda_dropped["level"])

# find all mid and senior
lda_dropped["level"] = np.where(["mid" in x for x in narr], "Mid-Senior level", lda_dropped["level"])
lda_dropped["level"] = np.where(["senior" in x for x in narr], "Mid-Senior level", lda_dropped["level"])

# find all executive
lda_dropped["level"] = np.where(["executive" in x for x in narr], "Executive", lda_dropped["level"])

lda_dropped.head()
```
